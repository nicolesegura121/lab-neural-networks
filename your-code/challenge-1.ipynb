```python
#Importing relevant libraries
import pandas as pd
import numpy as np 
```

```python
#Reading the data

df = pd.read_csv("tic-tac-toe.csv")
df
```

```python
#Data inspection

df.isnull().sum()# No missing data
df.shape # Data is 958 rows x 10 columns 
df.info() # all values are objects
df.columns #columns have an additional space in front. 

```

```python
#Data cleaning
df_clean= df.copy()
df_clean.columns = df_clean.columns.str.strip()
df_clean.columns
```

```python
df_clean = df_clean.apply(lambda x: x.str.strip())
df_clean["TL"].unique()
```

```python
#Converting values from categorical to numerical
# x will be 1 
# o will be 0
#True == 1
#False == 0

#Returning true or false as numeric values first: 
df_num = df_clean.replace({"True":1, "False":0,"x":1, "o":0, "b":2})
df_num
```

```python
#Target is the class, rest is feature
target = df_num["class"]
features = df_num.drop("class",axis=1)
```

```python
#Splitting the data: 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2,random_state=42)
```

```python
#Checking the shape of my train data
X_train.shape
```

```python
y_train.shape
```

*Defining the model:* 
1. Input shape: The features have 9 columns (766x9), so each combination should be equivalent to the target: 1 for win, 0 for loose or draw. 
2. Output is binary! there is no additional option if it is a draw. 


```python
#Model setup: 
import keras 
from keras import layers
from keras import ops
```

```python
#Creating a sequential model: 
from tensorflow.keras.models import Sequential
model = Sequential() #Layer by layer
```

```python
#Creating the layers for the model: 
from keras.optimizers import Adam
from tensorflow.keras.layers import Dense, Dropout, Input, Flatten
from tensorflow.keras.optimizers import RMSprop

model.add(Input(shape=(9,),dtype=int))  #  define the input shape
model.add(Flatten()) #flattens the inputs, it needs a 1D array
model.add(Dense(90, activation='relu', input_shape=(9,)))#Hidden layer
model.add(Dropout(0.2))#used to avoid overfitting, drops 20% of the neurons
model.add(Dense(30, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
#Modifying the learning rate
opt= tf.keras.optimizers.Adam(
    learning_rate=0.01,
    epsilon=1e-07,
)
model.compile(optimizer=opt,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#Saving the model 
model.save('tic-tac-toe.keras')
```
